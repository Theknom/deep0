```python
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")
    def hello(self):
        print("Hello " + self.name + "!")
    def goodbye(self):
        print("Goodbye " + self.name + "!")
```

```python
m = Man("Anya")
m.hello()
m.goodbye()
```

```python
import numpy as np

d = [[51, 55], [14, 19], [0, 4]]
X = np.array(d)
print(X)
Y = X.flatten()
print(Y)
Y[np.array([0, 2, 4])]  # ndarrayは[]の中にインデックス（ndarray）を入れることで、各値を取得できる
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos", linestyle="dashed")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin and cos")
# plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos", linestyle="dashed")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin and cos")
plt.legend() # 線の説明を表示
plt.grid() # グリッドを表示
plt.xlim(0, 10) # x軸の範囲を指定
plt.ylim(-1, 1) # y軸の範囲を指定
plt.show()
```
